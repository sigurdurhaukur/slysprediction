```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
data = pd.read_csv("accident-data.csv")

# Extract the date and amount of accidents columns
dates = data["Date"]

# Convert the dates to datetime objects
dates = pd.to_datetime(dates)
accidents = data["Amount of accidents"]
```

```python

decimal_places = 3

average_accidents = round(accidents.mean())
time_span = dates.max() - dates.min()

average_accidents_per_week = round(average_accidents / (time_span.days / 7), decimal_places)

print("Average accidents per month:", average_accidents)
print("Average accidents per week:", average_accidents_per_week)
print("for the last", time_span.days, "days")
print("for the last", time_span.days / 365.25, "years")
```

```python
# Plot the data
plt.figure(dpi=200)
plt.scatter(dates, accidents)
plt.xlabel("Year")
plt.ylabel("Amount of Accidents")
plt.title("Amount of Car Accidents over Time in Iceland")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing
plt.show()
```
