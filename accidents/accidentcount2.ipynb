```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('US-data.csv')
data = data.dropna()

data
```

```python
# count the number of accidents per degree celsius
accident_count = data.groupby('Temperature(C)').size()
accident_count
```

```python
date = data['Start_Time']

# convert date to datetime format
date = pd.to_datetime(date)

accident_count = data.groupby(date).size()
accident_count

plt.figure(dpi=200)
plt.scatter(accident_count.index, accident_count)
plt.xlabel('Year')
plt.ylabel('Amount of Accidents')
plt.title('Amount of Car Accidents over Time in the US')
plt.show()
```
