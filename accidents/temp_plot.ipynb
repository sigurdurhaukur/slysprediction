```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

# temperature agains accidents
data = pd.read_csv(
    "./hiti.csv",
    delimiter=",",
    names=["temperature", "amount of accidents"],
    skiprows=1,
)

data = data.sort_values(by=["temperature"])
data = data.drop(data[data["amount of accidents"] == 0].index)
temperature = data["temperature"]
amount_of_accidents = data["amount of accidents"]

plt.figure(dpi=200)
plt.title("Temperature against Amount of Accidents")
plt.scatter(temperature, amount_of_accidents)
plt.xlabel("Temperature (°C)")
plt.ylabel("Amount of accidents")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# temperature agains accidents
data = pd.read_csv(
    "./hiti.csv",
    delimiter=",",
    names=["temperature", "amount of accidents"],
    skiprows=1,
)

data = data.sort_values(by=["temperature"])
data = np.log(data)

data = data.drop(data[data["amount of accidents"] == 0].index)
temperature = data["temperature"]
amount_of_accidents = data["amount of accidents"]

plt.figure(dpi=200)
plt.title("Log of the Temperature against Amount of Accidents")
plt.scatter(temperature, amount_of_accidents)
plt.xlabel("Log of the Temperature (°C)")
plt.ylabel("Amount of accidents")
# plt.show()
plt.savefig("../figures/highres/log_temperature_against_amount_of_accidents.png")
```

```python

# wind speed against accidents
data = pd.read_csv(
    "./vindur.csv",
    delimiter=",",
    names=["wind speed", "amount of accidents"],
    skiprows=1,
)

# drop empty wind speed
data = data.dropna()

wind_speed = data["wind speed"]
amount_of_accidents = data["amount of accidents"]
# reverse the order of the data
wind_speed = wind_speed[::-1]
amount_of_accidents = amount_of_accidents[::-1]

plt.figure(dpi=200)
plt.title("Wind Speed against Amount of Accidents")
plt.scatter(wind_speed, amount_of_accidents)
plt.xlabel("Wind speed (m/s)")
plt.ylabel("Amount of accidents")
plt.show()
```
